```python
import pandas as pd
import seaborn as sns
sns.set()

from state_smoother import Smoother
```

```python
market_data = pd.read_csv(
    './_test/data/eve_market_data.csv',
    parse_dates=['record_date'], index_col='record_date'
)
```

```python
exp = Smoother(market_data, dims=3, horizon=14, verbose=True)
exp.learn(1000, learn_rate=1e-2)
```

```python
exp.plot_fit('Tritanium')
```

```python
pd.DataFrame(
    1 - exp._init_flow(exp.coefs['flow'], exp.horizon).detach().cpu().numpy(),
    columns=exp.columns,
)
```
